## Automate calculations for discrete variables

```python
def woe_discrete(df, discrete_variable_name , good_bad_variable_df):
    df = pd.concat([df[discrete_variable_name],good_bad_variable_df], axis = 1) #concatenate the variable of interest and the target columns
    df = pd.concat([df.groupby(df.columns.values[0], as_index=False)[df.columns.values[1]].count(), #groups by categories in the first column and count them
                    df.groupby(df.columns.values[0], as_index=False)[df.columns.values[1]].mean()], axis=1) #groups and gets the mean
    df = df.iloc[:,[0,1,3]] #keep only the first, second and fourth column
    df.columns = [df.columns.values[0],'n_obs','prop_good'] #names the columns
    df=df.copy()
    df['prop_n_obs'] = df['n_obs'] / df['n_obs'].sum() #gets the proportion of observation of a category
    df['n_good'] = df['prop_good']*df['n_obs'] #calculate the number of good borrowers
    df['n_bad'] = (1-df['prop_good'])*df['n_obs'] #calculate the number of bad 
    df['prop_n_good'] = df['n_good']/df['n_good'].sum() #proportion of good by bad
    df['prop_n_bad'] = df['n_bad']/df['n_bad'].sum() #proportion of bad by good
    df['WoE']=np.log(df['prop_n_good']/df['prop_n_bad']) #calculates weight of evidence according to formula
    df=df.sort_values(['WoE']) #sorts rows according to weight of evidence
    df=df.reset_index(drop=True) #resets index
    df['diff_prop_good']=df['prop_good'].diff().abs() #substracts two subsequent values in the column prop_good and gets the absolute value
    df['diff_WoE'] = df['WoE'].diff().abs() #substracts two subsequent values in the column WoE and gets the absolute value
    df['IV']=(df['prop_n_good']-df['prop_n_bad'])*df['WoE'] #calculate Information Value according to the formula
    df['IV']=df['IV'].sum() #sums all IV values to get the IV of a discrete value
    return df
```

```python
def plot_by_woe(df_WoE, rotation_of_labels=0):
    x=np.array(df_WoE.iloc[:,0].apply(str)) #matplotlib doesn't like dataframes and strings
    y=df_WoE['WoE']
    plt.figure(figsize=(18,6))
    plt.plot(x,y, marker = 'o', linestyle = '--', color = 'k')
    plt.xlabel(df_WoE.columns[0])
    plt.ylabel('Weight of evidence')
    plt.title(str('Weight of Evidence'+df_WoE.columns[0]))
    plt.xticks(rotation= rotation_of_labels)
```
